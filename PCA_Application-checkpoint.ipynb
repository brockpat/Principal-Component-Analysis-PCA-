### Goal
Here I provide an example for how to use PCA on stock returns. The Principal Components will capture some latent factors that we have to interpret based on their loadings

# Step 1) Obtain Data
For the sake of replication I use data from Yahoo Finance. Let's download the available data for the German DAX companies of the year 2024

```python
#%% Libraries
import yfinance as yf
import yahoo_fin.stock_info as si


import pandas as pd
import numpy as np

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt

import copy
```

```python
#%% Download Data
# Define the Tickers
dax_ticker = [
    "ADS.DE", "AIR.DE", "ALV.DE", 
    "BAS.DE", "BAYN.DE", "BEI.DE", "BMW.DE", "BNR.DE", 
    "CBK.DE", "CON.DE","1COV.DE", 
    "DTG.DE", "DBK.DE", "DB1.DE", "DHL.DE", "DTE.DE", 
    "EOAN.DE", "ENR.DE"
    "FRE.DE", 
    "HNR1.DE", "HEN3.DE", 
    "IFX.DE", 
    "MBG.DE", "MRK.DE", "MTX.DE", "MUV2.DE", 
    "P911.DE", "PAH3.DE", 
    "QIA.DE", 
    "RHM.DE","RWE.DE", 
    "SRT3.DE", "SAP.DE", "SIE.DE","SY1.DE", 
    "VNA.DE", "VOW3.DE", 
    "ZAL.DE"
]

start_date = "2024-01-01"
end_date = "2024-12-31"

# Download data from Yahoo Finance
df_raw = yf.download(dax_ticker, start=start_date, end=end_date)

#--- Download Names for Plots
# Dictionary to store ticker and company name
ticker_names = {}

# Loop through each ticker and get the company name
for ticker in dax_ticker:
    stock = yf.Ticker(ticker)
    company_name = stock.info.get('longName', 'Name not found') 
    ticker_names[ticker] = company_name
    
# Loop through the dictionary and replace 'Aktiengesellschaft' with 'AG'
for ticker, name in ticker_names.items():
    if 'Aktiengesellschaft' in name:
        ticker_names[ticker] = name.replace('Aktiengesellschaft', 'AG')

#Editing
ticker_names['MUV2.DE'] = 'Munich Re'
ticker_names['P911.DE'] = 'Porsche AG'
ticker_names['BMW.DE']  = 'BMW AG'
ticker_names['PAH3.DE'] = 'Porsche Holding SE'
```

## Data Processing
Next, I process the data and compute the returns and organise the data such that each stock is a feature (i.e. a column in the data frame) and each row is a time point (observation).

I also standardise the returns to mean zero and unit variance. This is helpful in many ways. Firstly, if one features varies more than the other solely due to its scale, then PCA would determine such a feature as more important only because of this. For example, body height has no natural unit and can be measured in many different units with vastly different scales. Luckily, returns have a natural unit since they are a relative scale.

Thus, when standardising the data here I assume that all variables (stocks) are predictive. However, if in reality the variables on a lower scale are not predictive, this will of course hurt model performance as noisy features will contribute more to the model's prediction. 

Lastly, PCA is usually applied to time-independent cases. However, stock returns are of course related over time, but in this case when applying a baseline PCA I assume that each new row is simply an ''ordinary'' observation. In other words, I'm neglecting any temporal effects. 

```python
#Reformat
df = (copy.deepcopy(df_raw)
         .unstack()
         .reset_index()
         .query("Price == 'Close'")
         .dropna()
         .drop('Price',axis= 1)
         .rename(columns = {0:'price'})
         )

#Compute Returns
df = (df
      .assign(ret = df.groupby('Ticker')['price'].transform(lambda x: x.pct_change()))
      )

#Finalise the DataFrame for PCA
df = (df
      .pivot(index='Date', columns='Ticker', values='ret')
      .dropna()
      #Standardize
      .pipe(lambda x: (x-x.mean())/x.std())
      )
```

## PCA Analysis
Now we are ready to conduct PCA. To determine the number of Principal Components, I simply check the Scree Plot. This is not the most sophisticated approach, but for this purpose it is sufficient. Based on the Scree Plot, I choose 3 Principal Components.

```python
#%% PCA Analysis

#Perform PCA
pca = PCA()
pca.fit(df)

# Scree Plot to determine Number of Principal Components
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, marker='o', linestyle='--', color='b')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.grid()
plt.show()
```

Next, I compute the portfolio weight of each stock for the Principal Components and plot these portfolio weights. This helps to interpret the Principal Components. 

```python
# Extract Loadings onto Principal components
df_loadings = pd.DataFrame(pca.components_, columns=df.columns)

#Generate Portfolio weights
df_portfolio = df_loadings.div(df_loadings.sum(axis=1), axis=0)

# Loop through the first 3 principal components
for i in range(0, 3):
    # Sort the portfolio weights for the i-th principal component in descending order
    sorted_weights = df_portfolio.iloc[i, :].sort_values(ascending=False)
    
    plt.figure(figsize=(11, 6))  # Adjust width (12 inches) and height (6 inches)
    
    
    # Create a bar plot
    plt.bar([ticker_names[ticker] for ticker in sorted_weights.index]
            , sorted_weights.values, color='b', alpha=0.7)
    
    # Add labels and title
    plt.xticks(rotation=90,fontsize=12)  # Rotate x-axis labels for better readability
    plt.ylabel("Portfolio Weight", fontsize = 14)  # Corrected from ylab to ylabel
    plt.title("Principal Component " + str(i + 1), fontsize = 16)
    
    # Show the plot
    plt.show()
```

The first principal component has positive weights for every stock. This is interpreted as the market factor so that the main driver of the variance is the overall market which is to be expected.

The second principal component (PC2) captures a sector-based contrast between car companies and insurance companies. Car companies have strongly negative weights, while insurance companies have strongly positive weights. This suggests that PC2 represents a latent factor that drives these two sectors in opposite directions, possibly related to interest rates, economic growth, or consumer confidence.

The third principal component (PC3) highlights a strong positive influence from the chemical and pharmaceutical industries. The lack of a clear pattern in the negative weights suggests that PC3 primarily captures a sector-specific factor rather than a contrasting relationship. Hence, the latent factor is one that benefits chemical and pharmaceutical companies such as innovation sin healthcare, regulatory developments or simply global demand for medical and chemical products. Therefore, the third principal component, due to them all being uncorrelated, is a sector-driven factor that isolates the performance of these industries from broader market trends.
